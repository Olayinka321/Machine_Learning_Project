#Oversampling

```python
import pandas as pd
from imblearn.over_sampling import RandomOverSampler
import matplotlib.pyplot as plt
import seaborn as sns

#Data PreProcessing 

#Loading the credit card dataset
df = pd.read_csv('creditcard.csv')

#Separation features and target variable
X = df.drop('class', axis=1)
y = df['class']

#Counting the number of instances in each class
class_counts_before = y.value_counts()

#Performing oversampling
oversampler = RandomOverSampler()
X_oversampled, y_oversampled = oversampler.fit_resample(X, y)

#Counting the number of instances in each class after oversampling
class_counts_after = pd.Series(y_oversampled).value_counts()

#Visualizing the class distribution before and after oversampling
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(y)
plt.title("Class Distribution Before Oversampling")
plt.xlabel("Class")
plt.ylabel("Count")
plt.subplot(1, 2, 2)
sns.countplot(y_oversampled)
plt.title("Class Distribution After Oversampling")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

#Printing the class counts before and after oversampling
print("Class Counts Before Oversampling:\n", class_counts_before)
print("\nClass Counts After Oversampling:\n", class_counts_after)
```

#Undersampling

```python
import pandas as pd
from imblearn.under_sampling import RandomUnderSampler
import matplotlib.pyplot as plt
import seaborn as sns

#Loading the credit card dataset
df = pd.read_csv('creditcard.csv')

#Separation features and target variable
X = df.drop('class', axis=1)
y = df['class']

#Counting the number of instances in each class
class_counts_before = y.value_counts()

#Perform undersampling
undersampler = RandomUnderSampler()
X_undersampled, y_undersampled = undersampler.fit_resample(X, y)

#Counting the number of instances in each class after undersampling
class_counts_after = pd.Series(y_undersampled).value_counts()

#Visualizing the class distribution before and after undersampling
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(y)
plt.title("Class Distribution Before Undersampling")
plt.xlabel("Class")
plt.ylabel("Count")
plt.subplot(1, 2, 2)
sns.countplot(y_undersampled)
plt.title("Class Distribution After Undersampling")
plt.xlabel("Class")
plt.ylabel("Count")
plt.tight_layout()
plt.show()

#Printing the class counts before and after undersampling
print("Class Counts Before Undersampling:\n", class_counts_before)
print("\nClass Counts After Undersampling:\n", class_counts_after)
```
